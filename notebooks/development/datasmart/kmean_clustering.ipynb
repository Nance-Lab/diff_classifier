```python
filename = 'Concessions.xlsx'
rfolder = 'datasmart'
aws.download_s3('{}/{}'.format(rfolder, filename), filename, bucket_name='ccurtis.data')
```

```python
class concessions(object):
    def __init__(self, item, category, price, profit, actual_profit, calories):
        self.item = item
        self.category = category
        self.price = price
        self.profit = profit
        self.actual_profit = actual_profit
        self.calories = calories

    def __str__(self):
        return("concessions object:\n"
               "  Item = {0}\n"
               "  Category = {1}\n"
               "  Price = {2}\n"
               "  Profit = {3}\n"
               "  Actual Profit = {4} \n"
               "  Calories = {5}"
               .format(self.item, self.category, self.price,
                       self.profit, self.actual_profit, self.calories))
```

```python
wb = open_workbook(filename)
counter = 0
for sheet in wb.sheets():
    if counter == 0:
        number_of_rows = sheet.nrows
        number_of_columns = sheet.ncols

        items = []

        rows = []
        for row in range(1, number_of_rows):
            values = []
            for col in range(number_of_columns):
                value  = (sheet.cell(row,col).value)
                try:
                    value = str(int(value))
                except ValueError:
                    pass
                finally:
                    values.append(value)
            item = concessions(*values)
            items.append(item)
    
    counter = counter + 1

for item in items:
    print("Accessing one single value (eg. Category): {0}".format(item.category))
```

```python
pditem = []
category = []
price = []
profit = []
actual_profit = []
calories = []

for item in items:
    pditem.append(item.item)
    category.append(item.category)
    price.append(item.price)
    profit.append(item.profit)
    actual_profit.append(item.actual_profit)
    calories.append(item.calories)
    
profit = []

for x in range(0, len(actual_profit)): profit.append(float(actual_profit[int(x)])/float(price[int(x)]))
```
