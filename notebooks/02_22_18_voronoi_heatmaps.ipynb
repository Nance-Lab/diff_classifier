```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi
import diff_classifier.heatmaps as hm

import os
import os.path as op
%matplotlib inline
```

```python
remote_folder = "01_18_Experiment/P1"
local_folder = os.getcwd()
ires = 512
prefix = 'P1_S1_L_0012'
frames = 651

merged_ft = pd.read_csv('features_{}.csv'.format(prefix))
points = np.zeros((merged_ft['X'].as_matrix().shape[0], 2))
points[:, 0] = merged_ft['X'].as_matrix()
points[:, 1] = merged_ft['Y'].as_matrix()
colors = merged_ft['asymmetry'].as_matrix()
```

```python
plt.figure(figsize = (12, 12))

#np.random.seed(1234)
# points = np.random.rand(15, 2)
# centers = np.random.rand(15, 2)

vor = Voronoi(points)

# plot
regions, vertices = hm.voronoi_finite_polygons_2d(vor)

# colorize
counter = 0
for region in regions:
    polygon = vertices[region]
    plt.fill(*zip(*polygon), alpha=0.4)
    centers[counter, :] = (np.mean(polygon[:, 0]), np.mean(polygon[:, 1]))
    counter = counter + 1

plt.plot(points[:,0], points[:,1], 'ko')
plt.plot(centers[:,0], centers[:,1], 'bo')
plt.xlim(0, 1)
plt.ylim(0, 1)
```

```python
for region in regions:
    polygon = vertices[region]
    x_points = np.mean(polygon[:, 0])
    y_points = np.mean(polygon[:, 1])
    print(x_points, y_points)
```
